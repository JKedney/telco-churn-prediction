# Telco Churn — Data Exploration

**Dataset:** 7,043 customers, 21 columns. Target: `Churn` (~73.5% No / 26.5% Yes — mildly imbalanced).

**Key data-quality finding:** `TotalCharges` is stored as text, not numeric. Converting it reveals 11 hidden blank values that `isnull()` missed — all customers with `tenure = 0` (brand-new sign-ups), so the blank is meaningful: they have no total charges yet.

**Cleaning plan (Session 5):** convert `TotalCharges` to numeric and fill those 11 blanks with 0.

```python
import pandas as pd

df = pd.read_csv("../data/WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["Churn"].value_counts(normalize=True)
```

```python
df["TotalCharges_numeric"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
df["TotalCharges_numeric"].isnull().sum()
```
